```python
import pandas as pd
from collections import namedtuple

import numpy as np
import pandas as pd

df = pd.read_csv('data/routes.csv', converters={'steps': pd.eval})
df.head()
```

```python
from geo import create_route, get_polyline_route

df['route'] = df['steps'].apply(create_route)

df['polyline_route'] = df['steps'].apply(get_polyline_route)

df.iloc[1]['polyline_route']
```

```python
import folium
import polyline

PARIS_CENTER = [48.854779, 2.350319]
map = folium.Map(location=PARIS_CENTER,
                zoom_start = 13) 

# trip_lines = [folium.PolyLine(locations=polyline.decode(line), weight=1.5, color='red', opacity=1) for line in df['polyline_route']]

route_lines = []
for route in df['polyline_route']:
    for l in route:
        line = folium.PolyLine(locations=polyline.decode(l), weight = 1.5, color = 'red', opacity = 0.8)
        route_lines.append(line)
        
for tl in route_lines: tl.add_to(map)

map
```
